```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('carprices.csv')
```

```python
data
```

```python
dummies = pd.get_dummies(data['Car Model']).astype(int)
```

```python
dummies
```

```python
merged = pd.concat([data,dummies],axis=1)
```

```python
merged
```

```python
data_dummies = merged.drop(['Car Model', 'BMW X5'], axis=1)
```

```python
data_dummies
```

```python
X = data_dummies.drop(['Sell Price($)'],axis=1)
```

```python
y = data_dummies['Sell Price($)']
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,1]])
```

```python
model.predict([[86000,7,0,0]])
```

```python
model.score(X,y)
```

```python
model_le.predict([[1,86000,7]])
```
