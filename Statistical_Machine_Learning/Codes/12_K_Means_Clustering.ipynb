```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns=iris['feature_names'])
df.head()
```

```python
data = df.drop(columns=['petal length (cm)','petal width (cm)'])
data.head()
```

```python
plt.scatter(data['sepal length (cm)'],data['sepal width (cm)'])
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_pred = km.fit_predict(data)
y_pred
```

```python
data['cluster'] = y_pred
data.head()
```

```python
data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
```

```python
plt.scatter(data1['sepal length (cm)'],data1['sepal width (cm)'],color='green')
plt.scatter(data2['sepal length (cm)'],data2['sepal width (cm)'],color='red')
plt.scatter(data3['sepal length (cm)'],data3['sepal width (cm)'],color='blue')
```

```python
scalar = MinMaxScaler()

data['sepal length (cm)'] = scalar.fit_transform(data[['sepal length (cm)']])
data['sepal width (cm)'] = scalar.fit_transform(data[['sepal width (cm)']])

data
```

```python
y_pred = km.fit_predict(data)
y_pred
```

```python
data['cluster'] = y_pred
data.head()
```

```python
km.cluster_centers_
```

```python
data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
```

```python
plt.scatter(data1['sepal length (cm)'], data1['sepal width (cm)'], color='green', label='Setosa')
plt.scatter(data2['sepal length (cm)'], data2['sepal width (cm)'], color='red', label='Versicolor')
plt.scatter(data3['sepal length (cm)'], data3['sepal width (cm)'], color='blue', label='Virginica')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker="*",label='Centroid')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(loc='upper right')
plt.title('Sepal Length vs Sepal Width by Iris Species')
plt.show()
```

### Elbow Method

```python
sse = []
for k in range(1,10):
  km = KMeans(n_clusters=k)
  km.fit(data)
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(range(1,10),sse)
plt.xlabel('K')
plt.ylabel('SSE')
plt.show()
```
